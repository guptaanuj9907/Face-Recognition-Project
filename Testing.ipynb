```python
import cv2
import numpy as np
import os
```

```python
cap=cv2.VideoCapture(0)
face_classifier=cv2.CascadeClassifier("C:/Users/CG-DTE/AppData/Local/Programs/Python/Python37/Scripts/OpenCV/Face Detection/haarcascade_frontalface_default.xml")
rec=cv2.face.LBPHFaceRecognizer_create() 
rec.read("C:/Users/CG-DTE/AppData/Local/Programs/Python/Python37/Scripts/OpenCV/face recognition project/recogniser.yml")
id=0
while True:
    ret,frame=cap.read()
    grey=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    face=face_classifier.detectMultiScale(grey,1.3,5)
    for fac in face:
        x,y,w,h=fac[:]
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),3)
        id,conf=rec.predict(grey[y:y+h,x:x+h])
        print(conf)
#         cv2.putText(frame,str(id),(x,x+w),cv2.FONT_HERSHEY_COMPLEX,10,(0,255,0))
        if id==1 and conf<120:
            cv2.putText(frame,'Obama'+','+str(np.around(conf))+'%',(x,y+h),cv2.FONT_HERSHEY_COMPLEX,2,(0,255,255),thickness=3)
        elif id==2 and conf<120:
            cv2.putText(frame,'Virat_Kohli'+','+str(conf)+'%',(x,y+h),cv2.FONT_HERSHEY_COMPLEX,2,(0,255,255),thickness=3)
        elif id==3 and conf<120:
            cv2.putText(frame,'Dhoni'+','+str(conf)+'%',(x,y+h),cv2.FONT_HERSHEY_COMPLEX,3,(0,255,0),thickness=3)
        else:
            cv2.putText(frame,'Hrithik',(x,y+h),cv2.FONT_HERSHEY_COMPLEX,3,(0,255,0),thickness=3)
    if cv2.waitKey(1)==13: 
            break   
    cv2.imshow("mine",frame)
    
cap.release()
cv2.destroyAllWindows()
```
